```python
df = pd.read_csv('../data/energy_dataset.csv')

print("✅ Датасетът е зареден!")
print(f"Размер: {df.shape[0]:,} реда и {df.shape[1]} колони")
print("\nПървите 5 реда:")
display(df.head())
```

```python
print("=== Основна информация за датасета ===")
print(df.info())

print("\n=== Статистическо описание ===")
display(df.describe())
```

```python
print("Липсващи стойности по колони:")
missing = df.isnull().sum().sort_values(ascending=False)
print(missing[missing > 0])
```

```python
print("Колони в датасета:")
print(df.columns.tolist())
```

```python
# Ключови колони, които ще ни интересуват най-много
key_columns = ['time', 'total load actual', 'total load forecast',
               'price actual', 'price day ahead']

print("Първите 10 реда от ключовите колони:")
display(df[key_columns].head(10))
```

```python
print("=== Статистика за Actual Load (реална консумация) ===")
print(df['total load actual'].describe())

print("\n=== Статистика за Цени (price actual) ===")
print(df['price actual'].describe())
```

```python
# Convert 'time' column to datetime with UTC handling
df['time'] = pd.to_datetime(df['time'], utc=True)

print("✅ Time column converted successfully (with UTC)!")
print("First 5 dates:\n", df['time'].head())
print("\nData period:", df['time'].min(), "to", df['time'].max())
print("Timezone:", df['time'].dt.tz)
```

```python
import matplotlib.pyplot as plt

# Plot overall energy consumption over time
plt.figure(figsize=(15, 6))
plt.plot(df['time'], df['total load actual'], label='Actual Load', color='blue', alpha=0.8)
plt.title('Total Energy Consumption Over Time (Spain)')
plt.xlabel('Time')
plt.ylabel('Load (MW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Average consumption by hour of the day
df['hour'] = df['time'].dt.hour

hourly_avg = df.groupby('hour')['total load actual'].mean()

plt.figure(figsize=(12, 6))
hourly_avg.plot(kind='bar', color='skyblue')
plt.title('Average Energy Consumption by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Load (MW)')
plt.grid(True)
plt.show()
```

```python
# Average consumption by day of week
df['day_of_week'] = df['time'].dt.day_name()

daily_avg = df.groupby('day_of_week')['total load actual'].mean()

plt.figure(figsize=(10, 5))
daily_avg.plot(kind='bar', color='coral')
plt.title('Average Energy Consumption by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Load (MW)')
plt.grid(True)
plt.show()
```

```python
# Extract month and create monthly average
df['month'] = df['time'].dt.month_name()

monthly_avg = df.groupby('month')['total load actual'].mean()

plt.figure(figsize=(12, 6))
monthly_avg.plot(kind='bar', color='darkblue')
plt.title('Average Energy Consumption by Month')
plt.xlabel('Month')
plt.ylabel('Average Load (MW)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(df['time'], df['total load actual'], label='Actual Load', color='blue', alpha=0.7)
plt.plot(df['time'], df['total load forecast'], label='Forecast Load', color='red', alpha=0.6)
plt.title('Actual vs Forecast Energy Load')
plt.xlabel('Time')
plt.ylabel('Load (MW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing_percent = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False)

missing_table = pd.concat([missing, missing_percent], axis=1, keys=['Missing Values', 'Percentage'])
print(missing_table[missing_table['Missing Values'] > 0])
```
