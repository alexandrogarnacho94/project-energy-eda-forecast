```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

# Load the dataset again
df = pd.read_csv('../data/energy_dataset.csv')

# Convert time column
df['time'] = pd.to_datetime(df['time'], utc=True)
df['time'] = df['time'].dt.tz_localize(None)   # Remove timezone for Prophet

print(" Dataset loaded successfully!")
print("Shape:", df.shape)
print("Time range:", df['time'].min(), "to", df['time'].max())
```

```python
# Prepare data for Prophet
prophet_df = df[['time', 'total load actual']].copy()
prophet_df = prophet_df.rename(columns={'time': 'ds', 'total load actual': 'y'})
prophet_df = prophet_df.dropna(subset=['y'])

# Train Prophet model
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False,
    seasonality_mode='multiplicative'
)

model.fit(prophet_df)

print(" Prophet model trained successfully!")
```

```python
# Generate future dates for the next 30 days (hourly)
future = model.make_future_dataframe(
    periods=30 * 24,      # 30 days * 24 hours
    freq='h'
)

# Make predictions
forecast = model.predict(future)

print(" Forecast generated successfully!")
print("Total forecast points:", len(forecast))
print("\nLast 5 predicted values:")
display(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

```python
# Show the forecast
fig = model.plot(forecast)
plt.title('Energy Consumption Forecast - Next 30 Days')
plt.xlabel('Date')
plt.ylabel('Load (MW)')
plt.grid(True)
plt.show()
```

```python
# Show trend and seasonal components
fig2 = model.plot_components(forecast)
plt.show()
```
